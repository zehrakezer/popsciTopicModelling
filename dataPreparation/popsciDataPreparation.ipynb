In the next step, I combined the title, bodytext and summary column into one column.

```python
import numpy as np
import re
import string
import seaborn as sns
```

```python
df["article"]= df.title+df.summary +df.bodytext
```

```python
df.drop(columns=["bodytext","summary","title"],inplace=True)
```

```python
df
```

removing unnecessary things like stop words, non-alphanumeric characters

```python
# Removing numeric words in comments
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', str(x))
# Remove punctuation and convert all words to lowercase
punc_lower = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x.lower()) 

df['article'] = df.article.map(alphanumeric).map(punc_lower)
df
```

remove special characters like ' or -

```python
import neattext.functions as nfx
df['article'] = df['article'].apply(nfx.remove_special_characters)
```

```python
df
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import nltk
nltk.download('stopwords')

stop = stopwords.words('english')

# Exclude stopwords with Python's list comprehension and pandas.DataFrame.apply.
df['article'] = df['article'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df
```

then I applied root extraction(lemmatization) techniques to reduce the words to their basic forms and remove the excess from the data

```python
from nltk import SnowballStemmer

sbs = SnowballStemmer(language='english')   
```

```python
## Getting to word roots
def stemmer(text):
    text = [sbs.stem(word) for word in text.split(' ') ]
    text = " ".join(text)
    return text

df.article = df.article.apply(stemmer)
```

```python
df
```

```python
from textblob import TextBlob
```

```python
#catches and fixes bugs

def correct(text):
    text = [str(TextBlob(word).correct()) for word in text.split(' ')]
    text = " ".join(text)
    return text

df.article = df.article.apply(correct)
```

```python
df
```

these stages helped us to get rid of different cases of the same word. 

In addition to these, I only took the nouns from the texts. Because it is not like sentiment analysis, so that words like good, better, worse do not cause pollution. I used post tagging(Parts of Speech Tagging) for this and wanted it to fetch the nouns.

```python
a=[]
for i in range(0,len(df)):
    sentences = nltk.sent_tokenize(str(df["article"][i]))#tokenize sentences

    for sentence in sentences:
        noun=[]
        for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
            if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS'):
                noun.append(word)
        a.append(' '.join(noun))

a
```

```python
len(a)
```

```python
df["article"] = a
```
