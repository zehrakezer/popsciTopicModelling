### WEB SCRAPING

I will extract data from https://www.popsci.com/

```python
import requests
from bs4 import BeautifulSoup as bts
from urllib.request import urlopen
import numpy as np
import pandas as pd
import re
from urllib.request import Request, urlopen
```

```python
def getAndParseURL(url):
    result = requests.get(url,headers={'user-agent':"Mozilla/5.0"})
    soup = bts(result.content, 'html.parser')
    return soup
```

First, let's scrape a list of which articles we will extract data from. Below is for the Gear category only, but I also scraped the extra tech, environment, science, health and DIY categories.

```python

LINK_LIST = []
for link in range(2,35):
    sayfa=getAndParseURL("https://www.popsci.com/category/gear/page/"+str(link)+"/")
    for lınk in sayfa.find_all("a",{"class":"PostItem-link"}):
      LINK_LIST.append(lınk["href"])
```

```python
# empty list to read list from a file
LINK_LIST = []

# open file and read the content in a list
with open(r"C:\Users\kezer\proje4\gearcategorilinklist.txt", 'r') as fp:
    for line in fp:
        # remove linebreak from a current name
        # linebreak is the last character of each line
        x = line[:-1]

        # add current item to the list
        LINK_LIST.append(x)

# display list
print(LINK_LIST)
```

```python
#define some lists
title=[]
summary=[]
categories=[]
bodytext=[]

for page in range(0,len(LINK_LIST)):
    #scraping the web page content
    result = requests.get(LINK_LIST[page],headers={'user-agent':"Mozilla/5.0"})
    sayfa = bts(result.text, 'html.parser')

    #get Title
    try:
        title.append(sayfa.find('h1', {'class':'Article-title'}).text)
    except:
        title.append(np.nan)
        
    #get the Summary
    try:
        summary.append(sayfa.find('p', {'class':'Article-excerpt'}).text.replace("\n",""))
    except:
        summary.append(np.nan)


    #get Body Text
    try:
        bodytext.append(sayfa.find("span",{"class":"ArticleReviewAuthor-publishedTime"}).findNext("p").text)
    except:
        bodytext.append(np.nan)


```

```python
#putting everything together
Articles = pd.DataFrame(list(zip(title, summary,bodytext)),columns = ["title", 'summary', 'bodytext'])
Articles
```

```python
category=[]
for i in range(len(Articles)):
    category.append("Gear")
```
